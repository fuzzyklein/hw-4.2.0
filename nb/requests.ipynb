```python
import urllib.request

try:
    rfc_number = int(input("What RFC are you looking for?"))
    url = f'http://www.rfc-editor.org/rfc/rfc{rfc_number}.txt'
    rfc_raw = urllib.request.urlopen(url).read()
    rfc = rfc_raw.decode()
    print(rfc)    
except (IndexError, ValueError):
    print("RFC must be an integer.")
    
```

```python
import requests

try:
    url = ('http://www.rfc-editor.org/rfc/rfc'
        + f'{int(input("What RFC are you looking for?"))}.txt')
    rfc = requests.get(url).text
    print(rfc)    
except (IndexError, ValueError):
    print("RFC must be an integer.")
    
```

```python
for k, v in requests.get(url).headers.items():
    print(f'{k}: {v}')
```
